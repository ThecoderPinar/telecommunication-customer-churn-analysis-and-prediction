# 🚀 Telecom Customer Churn Prediction Project 📊

## 🎯 Amaç ve Hedef:

Bu projenin temel amacı, telekomünikasyon müşterilerinin ayrılma durumlarını tahmin etmek ve bu tahminleri kullanarak müşteri kaybını azaltmaya yönelik stratejiler geliştirmektir. Projemizin ana hedefleri şunlardır:

1. Müşteri kaybını anlamak ve müşteri ayrılmasını etkileyen faktörleri belirlemek.
2. Müşteri kaybını tahmin etmek için doğru ve güvenilir bir model oluşturmak.
3. Müşteri kaybını azaltmaya yönelik stratejiler ve öneriler geliştirmek.

## 📋 Veri Seti Tanımı:

Bu projede kullanacağımız veri seti, telekomünikasyon müşterileriyle ilgili çeşitli öznitelikleri içerir. Veri setimizde toplamda 7043 satır ve 21 sütun bulunmaktadır. İşte veri setimizde bulunan sütunların detaylı bir açıklaması:

1. **customerID**: Müşteri kimliği
2. **gender**: Müşterinin cinsiyeti (Erkek veya Kadın)
3. **SeniorCitizen**: Müşterinin yaşlı bir vatandaş olup olmadığı (1: Evet, 0: Hayır)
4. **Partner**: Müşterinin bir partneri olup olmadığı (Evet, Hayır)
5. **Dependents**: Müşterinin bakmakla yükümlü olduğu kişilerin olup olmadığı (Evet, Hayır)
6. **tenure**: Müşterinin şirketle geçirdiği ay sayısı
7. **PhoneService**: Müşterinin telefon hizmeti olup olmadığı (Evet, Hayır)
8. **MultipleLines**: Müşterinin birden fazla hatta sahip olup olmadığı (Evet, Hayır, Telefon hizmeti yok)
9. **InternetService**: Müşterinin internet hizmet sağlayıcısı (DSL, Fiber optik, Yok)
10. **OnlineSecurity**: Müşterinin çevrimiçi güvenliğe sahip olup olmadığı (Evet, Hayır, İnternet hizmeti yok)
11. **Churn**: Müşterinin ayrılma durumu (Hedef değişkenimiz)

Bu veri setindeki sütunlar, müşteri kaybını tahmin etmek için kullanılacak olan temel öznitelikleri temsil eder. "Churn" sütunu hedef değişkenimizdir ve projemizin odak noktasıdır.

## 📈 Veri Analizi ve Keşif (EDA):

```python
# Aykırı değerleri tespit etmek için Z-Skorlarını hesaplayın
df_zscore = df[numeric_columns].apply(zscore)
```

```python
# Aykırı değerlerin varlığını belirlemek için Z-Skorlarını kullanarak eşik değeri belirleyin
outlier_threshold = 3
outliers = (np.abs(df_zscore) > outlier_threshold).any(axis=1)
```

```python
# Aykırı değerlere sahip gözlemleri veri setinden kaldırın
df_no_outliers = df[~outliers]
```

4. **En İyi Modelin Değerlendirilmesi**: En iyi parametreler ve bu parametrelerle elde edilen en iyi çapraz doğrulama skoru ekrana yazdırılır. Son olarak, en iyi model test seti üzerinde değerlendirilir ve test seti üzerindeki performansı hesaplanarak ekrana yazdırılır.

```python
# Veri setini hazırlama
encoder = OneHotEncoder()
y_train_encoded = encoder.fit_transform(y_train.reshape(-1, 1)).toarray()
y_test_encoded = encoder.transform(y_test.reshape(-1, 1)).toarray()
```

Bu kod parçasında, OneHotEncoder'ı varsayılan parametrelerle başlatıyoruz ve ardından .toarray() yöntemini kullanarak sonucu bir NumPy dizisine dönüştürüyoruz.

Şimdi, derin öğrenme modelimizi oluşturabiliriz.

```python
# Derin öğrenme modeli oluşturma
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(3, activation='softmax')
])
```

```python
# Modeli derleme
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Modeli eğitme
history = model.fit(X_train_scaled, y_train_encoded, epochs=50, batch_size=8, validation_split=0.2)
```

Bu kod parçasında, Sequential modeli ile bir katmanlar dizisi oluşturduk. Modelimizde iki gizli katman ve bir çıkış katmanı bulunmaktadır. Son olarak, modeli derledik ve eğittik.

Eğitim süreci boyunca doğruluk ve kayıp (loss) değerlerini görselleştirmek için aşağıdaki kodu kullanabiliriz:

```python
# Eğitim sürecini görselleştirme
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
